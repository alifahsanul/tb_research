```python
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from keras.models import load_model
from keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array
import numpy as np
import cv2
import matplotlib.pyplot as plt
import os
import my_config
import grad_cam_lib
os.getcwd()
```

```python
model_base_arch_selected = my_config.model_base_arch_selected
print(model_base_arch_selected)
img_data_root_path = '../data_prep/data_for_modelling'
image_size = my_config.image_size
print(image_size)
```

```python
from keras.models import load_model
model = load_model(os.path.join('model_result', model_base_arch_selected, '006-0.8939.hdf5'))
model.summary()
```

```python
def my_predict(model, file):
    x = load_img(file, target_size=(image_size))
    x = img_to_array(x)
    x = x / 255.0
    x = np.expand_dims(x, axis=0)
    array = model.predict(x)
    assert array.ndim == 2
    assert array.shape[0] == 1
    prob_arr = array[0]
    answer = np.argmax(prob_arr)
    return answer, prob_arr

def get_random_data(data_df, split, n_sample=1):
    df = data_df[data_df['split']==split]
    n_sample = min(n_sample, len(df))
    df = df.sample(n=n_sample)
    result_list = []
    for ind, row in df.iterrows():
        class_type = 'tb' if row['is_sick'] else 'normal'
        filename = row['img_name'] + '.png'
        source_file_path = os.path.join(img_data_root_path, split, class_type, filename)
        image = load_img(source_file_path, target_size=image_size)
        predicted_class, prob_arr = my_predict(model, source_file_path)
        img_dict = row.to_dict()
        img_dict['image'] = image
        img_dict['filepath'] = source_file_path
        img_dict['predicted_class'] = predicted_class
        img_dict['prob_arr'] = np.round(prob_arr, 2)
        result_list.append(img_dict)
    return result_list

data_df = pd.read_pickle(os.path.join(img_data_root_path, 'data_df.pkl'))
sample_img_list = get_random_data(data_df, 'test', 7)
```

```python
last_layer_index = -4
last_conv_layer_name = model.layers[last_layer_index].name
print(last_conv_layer_name)
classifier_layer_names = [x.name for x in model.layers[last_layer_index+1:]]
print(classifier_layer_names)
```

```python
for i, el in enumerate(sample_img_list):
    grad_cam_lib.perform_grad_cam(el, model, last_conv_layer_name,
                                  classifier_layer_names, image_size)
    print('\n\n')
```
